```python
import numpy as np
import os


file_dir = 'C:/Users/BVer/are_you_ok/dataset/' #seq 파일을 dataset 폴더에 넣어주세요
count = 0
for name in os.listdir(file_dir):
    state = 0
    if not len(np.load(file_dir + name)) == 0:
        npy_file = np.load(file_dir + name)

        for k in range(0, len(npy_file)):
            for j in range(30):
                for i in range(len(npy_file[k][j])):
                    t = npy_file[k][j][i]
                    if np.isnan(t):
                        state = 1
                        count += 1
                        break
                if state == 1:
                    break
            if state == 1:
                print(name)
                os.rename(file_dir + name, file_dir + 'nan_seq_' + str(count) + '.npy')
                break
        
print(count)
```
